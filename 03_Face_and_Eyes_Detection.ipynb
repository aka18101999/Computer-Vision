### Import Libraries

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

### Create Classifiers

```python
face_classifier=cv2.CascadeClassifier('Haarcascades/haarcascade_frontalface_default.xml')
eye_classifier=cv2.CascadeClassifier('Haarcascades/haarcascade_eye.xml')
```

### Open Image

```python
img=cv2.imread('Images/IMG-20190808-WA0007.jpg')
```

### Fix Image

```python
fix_img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

### Face Classifier

```python
faces=face_classifier.detectMultiScale(fix_img,1.3,5)
```

### If No Face Detected

```python
if faces is ():
    print('no faces found')
```

### It's a Kind of Magic

```python
def detect_face_eyes(fix_img):
    face_reacts=face_classifier.detectMultiScale(fix_img)
    for(x,y,w,h) in face_reacts:
        cv2.rectangle(fix_img,
                      (x,y),
                      (x+w,y+h),
                      (255,0,0),
                      10)
        eyes_reacts=eye_classifier.detectMultiScale(fix_img)
        for(ix,iy,iw,ih) in eyes_reacts:
            cv2.rectangle(fix_img,
                      (ix,iy),
                      (ix+iw,iy+ih),
                      (255,255,255),
                      10)
    return fix_img
```

```python
result= detect_face_eyes(fix_img)
plt.imshow(result)
```
